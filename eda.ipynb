```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.dtypes
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
df.dtypes
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values()
region_sales.plot.barh( title = 'Region_wise Sales')
plt.xlabel("Sales")
plt.ylabel("Region")
```

```python
category_wise_profit = df.groupby('Category')['Profit'].sum()
category_wise_profit.plot.pie(title = 'Category-wise Profit Share',autopct = '%.2f%%',startangle=140,
    explode=[0.05, 0.05, 0.05])
plt.ylabel('')
```

```python
monthly_trend = df.groupby('Order Month')['Sales'].sum()
monthly_trend.plot(title = 'monthly_trend',figsize = (12,5),marker='o', color='green')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
```

```python
sns.heatmap(df[['Sales','Profit','Quantity','Discount']].corr(),annot = True,fmt='.2f',
    cmap='coolwarm',
    square=True
)
plt.title("Correlation Between Key Metrics")
```

```python
pivot = df.pivot_table(index='Region', columns='Segment', values='Profit', aggfunc='sum')
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Profit by Segment and Region")
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10)
top_products.plot.barh(title = 'Top 10 Products by Sales',color = 'orange')
plt.xlabel("Sales") 
```
